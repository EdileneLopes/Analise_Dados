# Web Scraping Beautifulsoup de músicas

```python
# Importando as bibliotecas
#urllib auxiliará a fazer as requisições ao site, e a BeautifulSoup que fará a extração de forma fácil
import pandas as pd
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
from bs4 import BeautifulSoup
```

```python
# URL base
url = 'https://www.imdb.com/chart/top?ref_=nv_mv_250'
```

```python
# Tratando erros da url
try:
    html = urlopen(url)
except HTTPError as e:
# Erros HTTP
    print(e)
except URLError as e:
    # URL errada
    print('A página não foi encontrada!')
```

```python
# Usando o bs4
bs = BeautifulSoup(html, 'lxml')

# Imprimindo o título da página
filmes = bs.select('.lister-list tr')
```

```python
#a estrutura dos títulos e selecionar o parâmetro title
#Ainda na classe td com classe titleColumn, selecionar o span que é o ano do filme
#selecionar a classe imdbRating e pegar o texto da tag strong dentro deste td
```

```python
print(filmes)
```

```python
titulos = []
diretores_escritores = []
anos = []
avaliacoes = []

for filme in filmes:
    titulos.append(filme.find('td', class_='titleColumn').find('a').get_text())
    diretores_escritores.append(filme.find('td', class_='titleColumn').find('a')['title'])
    anos.append(filme.find('td', class_='titleColumn').find('span').get_text()[1:5])
    avaliacoes.append(filme.find('td', class_='imdbRating').find('strong').get_text())
```

```python
print(titulos)
```

```python
# Conferindo o nNúmero de titulos
print('Foram encontradas {} ocorrências de titulos'.format(len(titulos)))
```

### Criando o Dataset

```python
df = pd.DataFrame({
        "titulos": titulos, 
        "anos": anos, 
        "avaliacoes": avaliacoes, 
        "diretores_escritores":diretores_escritores
    })

df.head()
```

### Salvando em um arquivo csv

```python
df.to_csv('imdb_melhores_filmes.csv')
```
